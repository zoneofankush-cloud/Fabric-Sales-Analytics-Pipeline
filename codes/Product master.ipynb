### 🛠️ Step 1: Denormalizing Product Hierarchy (Snowflake to Star Schema)
**Objective:** Flatten the highly normalized product lookup tables (Product, Subcategory, Category) into a single, unified `product_master` dimension table. 

**Architectural Benefit:** This eliminates the "Snowflake" schema, reducing the number of hops the Power BI VertiPaq engine needs to make during filtering, resulting in significantly faster dashboard loading times.


## Do not run this cell again, this can create duplicates to master_product

```python
%%sql
CREATE TABLE ankush_schema.product_master
(SELECT
pl.ProductKey,
pl.ProductName,
pc.CategoryName,
psc.SubcategoryName,
pl.ModelName,
pl.ProductColor,
pl.ProductSize,
pl.ProductStyle,
pl.ProductSKU,
pl.ProductCost,
pl.ProductPrice
FROM ankush_schema.`product lookup` pl
 left join ankush_schema.`product subcategories lookup` psc
 on psc.ProductSubcategoryKey = pl.ProductSubcategoryKey

 left JOIN ankush_schema.product_category_lookup pc
 on pc.ProductCategoryKey = psc.ProductCategoryKey)
```
